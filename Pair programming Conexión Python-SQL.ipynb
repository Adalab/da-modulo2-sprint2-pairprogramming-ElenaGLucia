#### 2. Crear la conexión a la base de datos Northwind que devuelva un mensaje de error si algo no ocurre cómo debería

```python
try:
  cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='localhost',
                              database='northwind')

except mysql.connector.Error as err:

    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR: # si es un error con la contraseña
        print("Something is wrong with your user name or password")
  
    elif err.errno == errorcode.ER_BAD_DB_ERROR: # si el error se debe a que la bdd no exista
        print("Database does not exist")
    
    else:
        print(err) # si no es por ninguno de los errores anteriores, printeame cual es el error que estoy teniendo en mi conexión
```

#### 4. Seleccionar todo de la tabla "customers" donde "country" sea "USA" o "Germany". Almacenar los resultados en una variable

```python
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database='northwind')

mycursor = cnx.cursor() # iniciamos el cursor

mycursor.execute("SELECT * FROM customers WHERE country = 'Germany' OR 'USA' ") # ejecutamos la query

myresult = mycursor.fetchall() # que nos saque todos los resultados de la variable

for x in myresult:
    print(x)
```
